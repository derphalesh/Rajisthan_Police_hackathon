```python
import pickle

# Load the model from the pickle file
with open('model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
csv_file_path = 'test.csv'

# Specify the row number you want to read (e.g., row 0 for the first row)
row_number_to_read = 0

# Read only one row into a DataFrame
df = pd.read_csv(csv_file_path)
# df = pd.read_csv(csv_file_path, nrows=row_number_to_read + 1)
```

```python
df
```

```python
columns_to_be_dropped = ['TransactionID','MerchantID','CustomerID','Name', 'Age', 'Address']
df = df.drop(columns_to_be_dropped, axis=1)
```

```python
df['Timestamp1'] = pd.to_datetime(df['Timestamp'])
df['Hour'] = df['Timestamp1'].dt.hour
df['LastLogin'] = pd.to_datetime(df['LastLogin'])
df['gap'] = (df['Timestamp1'] - df['LastLogin']).dt.days.abs()
```

```python
df = df.drop(['FraudIndicator','Timestamp','Timestamp1','LastLogin'],axis=1)
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame and 'Category' is the column to encode
all_categories = ['Online', 'Other', 'Travel', 'Food', 'Retail']  # List of all possible categories

for category in all_categories:
    new_column_name = f'Category_{category}'
    df[new_column_name] = (df['Category'] == category).astype(int)

# Drop the original 'Category' column
df.drop('Category', axis=1, inplace=True)
```

```python
df
```

```python
fix_order = ['TransactionAmount','AnomalyScore','Amount','AccountBalance','SuspiciousFlag','Hour','gap','Category_Food','Category_Online','Category_Other','Category_Retail','Category_Travel']
```

```python
df = df.reindex(columns=fix_order)
```

```python
predictions = loaded_model.predict(df)
predictions
```

```python
probabilities = loaded_model.predict_proba(df)

# Assuming '1' is the label for fraud
fraud_probabilities = probabilities[:, 1]

# You can set a threshold (e.g., 0.5) to classify a transaction as fraudulent
threshold = 0.5
predicted_fraudulent = fraud_probabilities > threshold
predicted_fraudulent
```
